```python
import pandas as pd

evaluations_df = pd.read_csv('soap_note_evaluations.csv')
evaluations_df['hallucination'] = evaluations_df['hallucination'].apply(lambda x: 1 if x == 'Yes' else 0)
evaluations_df.head()
```

```python
evaluations_df.describe()
```

```python
grouped_evaluations = evaluations_df.groupby(['model_used', 'model_size']).agg({
    'completeness': ['mean', 'std', 'min', 'max'],
    'conciseness': ['mean', 'std', 'min', 'max'],
    'hallucination': ['mean', 'std', 'min', 'max'],
}).reset_index()

grouped_evaluations
```
